```python
from sklearn.preprocessing import normalize
x = normalize(x)
#x = normalize(x, axis = 0, norm = 'max')
```

```python
from sklearn.cross_validation import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
x_test, x_val, y_test, y_val= train_test_split(x_test, y_test, test_size=0.5, random_state=1)
```

```python
print(y_train)
```

```python
print(y_test)
```

```python
#y reshape하기
y = np.reshape(y,(569,1))
y_train = np.reshape(y_train,(398,1))
y_test = np.reshape(y_test, (85,1))
y_val = np.reshape(y_val, (86,1))
```

```python
sess=tf.Session()
sess.run(tf.global_variables_initializer())
```

```python
##빈벡터만들기
new_x =tf.placeholder(tf.float32, shape=[None, 30])
new_y =tf.placeholder(tf.float32, shape=[None, 1])
w1 =tf.Variable(tf.random_normal([30,16]), name='weight1')
b1 =tf.Variable(tf.random_normal([16]), name='bias1')
w2 =tf.Variable(tf.random_normal([16,16]), name='weight2')
b2 =tf.Variable(tf.random_normal([16]), name='bias2')
w3 =tf.Variable(tf.random_normal([16,16]), name='weight3')
b3 =tf.Variable(tf.random_normal([16]), name='bias3')
w4 =tf.Variable(tf.random_normal([16,1]), name='weight4')
b4 =tf.Variable(tf.random_normal([1]), name='bias4')

 
L1 = tf.nn.relu(tf.matmul(new_x,w1)+b1)
L2 = tf.nn.relu(tf.matmul(L1,w2)+b2)
L3 = tf.nn.relu(tf.matmul(L2,w3)+b3)
hypothesis =tf.sigmoid(tf.matmul(L3,w4) +b4)


cost = -tf.reduce_mean(new_y*tf.log(hypothesis) + (1-new_y)*tf.log(1-hypothesis))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-3).minimize(cost)


# Accuracy computation
# True if hypothesis>0.5 else False
predicted = tf.cast(hypothesis > 0.5, dtype=tf.float32)
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted, new_y), dtype=tf.float32))
```

```python
# Launch graph
with tf.Session() as sess:
    
    merged_summary = tf.summary.merge_all()
    writer = tf.summary.FileWriter("./logs/logs_t2")
    writer.add_graph(sess.graph)
    
    # Initialize TensorFlow variables
    sess.run(tf.global_variables_initializer())
    
    for step in range(118600):
        cost_val, _ = sess.run([cost, optimizer], feed_dict={new_x: x_train, new_y: y_train})
        if step % 100 == 0:
            print(step, cost_val)
    
    
    # Accuracy report : test_set
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={new_x: x_train, new_y: y_train})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\nTrain Accuracy: ", a)
            
    # Accuracy report : val_set
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={new_x: x_val, new_y: y_val})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\n Validation Accuracy: ", a)
    
     # Accuracy report : val_set
    h, c, a = sess.run([hypothesis, predicted, accuracy], feed_dict={new_x: x_test, new_y: y_test})
    print("\nHypothesis: ", h, "\nCorrect: ", c, "\n test Accuracy: ", a)
```
